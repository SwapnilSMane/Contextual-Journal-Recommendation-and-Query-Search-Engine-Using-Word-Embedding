```python
import glob, os
import joblib

import pandas as pd

from data_genration import data_prepare
```

```python
folder = '/home/swapnil/Documents/ML1/Project/Implementation/journals'
subfolders = [ f.path for f in os.scandir(folder) if f.is_dir() ]
```

```python
def readfiles():
    os.chdir(path)
    pdfs = []
    for file in glob.glob("*.pdf"):
        pdfs.append(file)
    return pdfs
```

```python
journals_df = []
journals_title = []
cls = data_prepare()
for p in subfolders:
    print(p)
    path = p
    files = readfiles()
    df_all=[]
    for file in files:
        
        df_all.append(cls.extract_text(file))
        journals_title.append(p.split('/')[-1])
        
    temp = pd.concat(df_all)
    journals_df.append(temp)
```

```python
df = pd.concat(journals_df)
df.reset_index(drop=True,inplace=True)
df['Journal'] = journals_title
```
